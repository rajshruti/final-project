```python
# Import dependencies

import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Read the file containing dataset

file = pd.read_csv("../Data/MLdataset.csv")
```

```python
file["TotalTime"] = file["TotalTime"].round(1)
```

```python
file.head()
```

# Applying One-Hot Encoding: TotalTime

```python
# Step 0: Reformat data
    
data = file.values
#X = data[:, 0:3]
X = data[:, 0]
y = data[:, 3]
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set

label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 20)
```

# Logistic Regression (Univariate: TotalTime)

```python
#  Split our data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
X_train = X_train.reshape(-1,1)

X_test = X_test.reshape(-1,1)
```

```python
# Create a Logistic Regression Model

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict the class of the new data point

import numpy as np
new_data = np.array([[567.4]])
predictions = classifier.predict(new_data)
print("Classes are either 0 (Active/Graduated) or 1 (Dropout)")
print(f"The new point was classified as: {predictions}")
```
